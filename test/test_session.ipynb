```python
import sys
sys.path.insert(1,'../../ipage/scripts/')
import ipage
```

```python
de_genes, de_profile = ipage.read_expression_file('bladder.exp', sep='\t')
```

```python
annotation_name = 'msig_db_gs'
de_gene_format = 'refseq'
annotation_gene_format = 'gs'
output_name = 'stdout'
annotation_dir = '../annotation_dir'
ipage_hits = ipage.ipage(de_genes, de_profile, annotation_name, output_name=output_name, de_ft=de_gene_format, ann_ft=annotation_gene_format, annotation_dir=annotation_dir)
```
